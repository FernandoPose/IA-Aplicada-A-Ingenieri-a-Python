## Información del dataset y contexto (lo mismo de nuevo, para comparar...)

**Conjunto de datos de calidad del aire para clasificación ADL (*Activities of Daily Living*)**

`Gambi, Ennio (2020), “Air Quality dataset for ADL classification”, Mendeley Data, V1, doi: 10.17632/kn3x9rz3kd.1`

La variación en la concentración de gas interior a lo largo del tiempo se monitorea y los datos se almacenan para usar esta información para evaluar el tipo de actividad realizada en la habitación. Gracias al uso de inteligencia artificial, se evitó un enfoque cuantitativo para determinar la concentración de gas, que habría requerido una cuidadosa calibración de los sensores. El conjunto de datos contiene los valores adquiridos por una matriz de 6 sensores de bajo costo en instantes sucesivos de tiempo, y los valores almacenados están asociados con la acción particular que los generó. A través de un procesamiento de datos apropiado, basado en algoritmos de aprendizaje automático, después de una fase de capacitación inicial, es posible reconocer las acciones que se llevan a cabo dentro del hogar. La presencia de productos químicos en el aire se determina a través de una serie de sensores de gas electroquímicos que se han seleccionado en función de las especificaciones técnicas establecidas sobre la capacidad de detectar clases de compuestos. El conjunto de sensores se puede agrupar en dos categorías principales:
- Sensores MQ ( MQ2, MQ9, MQ135, MQ137, MQ138 ) que tienen gran sensibilidad, baja latencia y bajo costo; cada sensor puede responder a diferentes gases;
- Sensor analógico de gas CO2 ( MG-811 ) que tiene una excelente sensibilidad al dióxido de carbono y apenas se ve afectado por la temperatura y la humedad del aire.

El conjunto de datos contiene 1845 muestras recolectadas que describen 4 situaciones objetivo:

1. Situación normal - Actividad: aire limpio, una persona que duerme o estudia
o descansa - Muestras: 595.
2. Preparación de comidas - Actividades: cocinar carne o pasta, verduras fritas. Una o dos personas en la habitación, circulación de aire forzado - Muestras: 515.
3. Presencia de humo - Actividad: quemar papel y madera por un corto período de tiempo en una habitación con ventanas y puertas cerradas - Muestras: 195.
4. Limpieza - Actividad: uso de detergentes en aerosol y líquidos con amoníaco y / o alcohol. La circulación de aire forzado se puede activar o desactivar - Muestras: 540.

Cada muestra está compuesta por 7 valores; los primeros seis valores son las salidas del sensor, mientras que el último es el índice de la acción que generó los valores adquiridos por los sensores. Las cuatro situaciones diferentes están asociadas con una composición del aire bastante diferente, teniendo en cuenta que cualquier actividad produce sustancias químicas (respiración humana, exhalaciones de procesos metabólicos, liberación de volátiles por combustión y/o oxidación, evaporación de detergentes domésticos, etc.).

## ***Setup***

Esta función inicializa el entorno de formación y crea el proceso de transformación. Se debe llamar a la función de configuración antes de ejecutar cualquier otra función en PyCaret. Solo tiene dos parámetros requeridos, es decir, datos y objetivo. Todos los demás parámetros son opcionales.

De todas maneras, pueden establecerse aquí todas las configuraciones necesarias para el tratamiento de los datos. Es decir, lo que surja cómo necesidad luego del Análisis Exploratorio de Datos.

```python
from pycaret.classification import *
s = setup(dataset, target = 'Situacion', session_id = 123)
```

Una vez que la configuración se ha ejecutado exitosamente, muestra la cuadrícula de información que contiene información del nivel del experimento.

- **Session id:** un número pseudoaleatorio distribuido como semilla en todas las funciones para su posterior reproducibilidad. Si no se pasa ningún session_id, se genera automáticamente un número aleatorio que se distribuye a todas las funciones.

- **Target type:** Binario, Multiclase o Regresión. El tipo de objetivo se detecta automáticamente.

- **Target mapping:** Cuando la variable de destino es de tipo cadena (es decir, 'Sí' o 'No') en lugar de 1 o 0, codifica automáticamente la etiqueta en 1 y 0 y muestra la asignación (0: No, 1: Sí) como referencia. En este ejemplo, notar que a pesar de estar como enteros las categorías, las reasigna para iniciar en 0 (algunos algoritmos así lo requieren).

- **Original data shape:** Forma de los datos originales antes de cualquier transformación.

- **Transformed train set shape:** Forma del conjunto de entrenamiento transformado

- **Transformed test set shape:** Forma del conjunto de pruebas transformado

- **Numeric features:** El número de características consideradas numéricas.

- **Categorical features:** El número de características consideradas categóricas.

```python
# Si se necesitan ver todas las configuraciones...
help(setup)
```

En el **setup** se configuran varias cosas. Para acceder a todas las variables creadas por la función de configuración, como el conjunto de datos transformado, random_state, etc., se puede utilizar la función **get_config()**. Abajo la usamos...

### Ejemplos para el preprocesamiento:

**Básicos:**
* normalize=True: Normaliza el conjunto de datos antes de entrenar el modelo.
* transformation=True: Aplica transformaciones automáticas como la conversión de variables categóricas.

**Avanzados:**
* preprocess=True: Activa las funciones de preprocesamiento automático.
* outliers=True: Identifica y trata automáticamente los valores atípicos.
* feature_interaction=True: Crea automáticamente interacciones entre características.

**Balanceo de Clases:**
* fix_imbalance=True: Aplica técnicas de remuestreo para abordar el desbalanceo de clases. Por defecto SMOTE (*Synthetic Minority Over-sampling Technique*).

**Transformaciones y Normalización:**
* transformation_method: Especifica el método de transformación.
* normalize_method: Define el método de normalización.

Y mucho más: PCA, descomposición de variables, selección de características, análisis de multicolinearidad, etc.

Todas las transformaciones de preprocesamiento se aplican dentro de la función de configuración. PyCaret proporciona más de 25 transformaciones de preprocesamiento diferentes. Más info en: [Data Preprocessing](https://pycaret.gitbook.io/docs/get-started/preprocessing)

## ***Compare Models***

Esta función entrena y evalúa el rendimiento de todos los estimadores disponibles en la biblioteca de modelos mediante validación cruzada. El resultado de esta función es una cuadrícula de puntuación con puntuaciones medias validadas de forma cruzada. Se puede acceder a las métricas evaluadas durante el CV mediante la función **get_metrics**. Las métricas personalizadas se pueden agregar o eliminar usando las funciones **add_metric** y **remove_metric**.

```python
# Comparar modelos de referencia (baseline), devuelve el mejorcito, mirando el Accuracy
best = compare_models()
```

**compare_models** utiliza de forma predeterminada todos los estimadores de la biblioteca de modelos (todos excepto los modelos con Turbo=False). Para ver todos los modelos disponibles podemos usar la función modelos()

```python
models()
```

Podemos usar el parámetro de inclusión y exclusión en compare_models para entrenar solo el modelo seleccionado o excluir modelos específicos del entrenamiento pasando la identificación del modelo en el parámetro de exclusión.

```python
# Por ejemplo: comparamos los modelos basados en árboles solamente
compare_tree_models = compare_models(include = ['dt', 'rf', 'et', 'gbc', 'xgboost', 'lightgbm', 'catboost'])
```

De forma predeterminada, compare_models devuelve el modelo único con mejor rendimiento según la métrica definida en el parámetro de clasificación. Podemos especificar la cantidad de modelos a devolver y el criterio basado en lo que necesitamos. Ejemplo: 3 modelos basados ​​en Recall

```python
best_recall_models_top3 = compare_models(sort = 'Recall', n_select = 3)
```

```python
# Los 3 mejores, ordenados por la métrica seleccionada. Por ejemplo: podriamos automatizar algún meta modelo con esto :)
best_recall_models_top3
```
